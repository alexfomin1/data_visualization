```python
import seaborn as sns
import pandas as pd
```

```python
df_tips = sns.load_dataset('tips')
df_tips
```

```python
df_tips.groupby(by=['day', 'time'])\
    .mean()\
    .fillna(0)\
    .style.format('{:.2f}').background_gradient(axis=0)
```

```python
df_tips.groupby(by=['day', 'time'])\
    .mean()\
    .fillna(0)\
    .style.format('{:.2f}').bar(axis=0, width=50, align='zero')
```

```python
len(df_tips.groupby('sex'))
```

```python
df_tips.groupby('sex').mean()
```

```python
df_grouped = df_tips.groupby('sex')
```

```python
df_grouped.get_group('Female')
```

```python
df_grouped.get_group('Female').mean(numeric_only=True)
```

```python
df_grouped.total_bill.sum()
```

```python
df_grouped.get_group('Female').total_bill.sum()
```

```python
df_tips.groupby(by=['day', 'smoker']).sum()
```

```python
df_tips.pivot_table(index='day', columns='smoker', aggfunc='sum')
```

```python
dfres = df_tips.pivot_table(index='day', columns='smoker', values='total_bill', aggfunc=['sum', 'mean'])
```

```python
dfres.style.background_gradient()
```

```python
df_tips.pivot_table(
    index='sex',
    columns='time',
    aggfunc=['sum', 'mean'],
    values='total_bill'
)
```

```python
df_tips.pivot_table(
    index='sex',
    columns='time',
    aggfunc=['sum', 'mean'],
    values=['total_bill', 'tip']
)
```

```python
df_tips.pivot_table(
    index=['day', 'smoker'],
    columns='time',
    aggfunc=['sum', 'mean'],
    values=['total_bill', 'tip']
).fillna(0)
```

```python
df_tips.pivot_table(index=['day', 'smoker'],
                    columns=['time', 'sex'],
                    aggfunc=['sum', 'mean'],
                    values=['total_bill', 'tip']
                    )
```
